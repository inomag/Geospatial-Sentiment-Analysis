```python
import pandas as pd
import folium
import math
```

```python
def getColor(x):
    R = math.floor((255*x)/100)
    G = math.floor((255*(100-x))/100)
    B = 0
    return '#%02x%02x%02x' % (R, G, B)
```

```python
df = pd.read_csv('./final datasets/Guwahati_Data.csv')

lat = df['Latitude']
lon = df['Longitude']

# Get nodes extent
margin = 0.01 # buffer to add to the range
lat_min = min(lat) - margin
lat_max = max(lat) + margin
lon_min = min(lon) - margin
lon_max = max(lon) + margin

# Get centre coordinates for map
mid_lat,mid_lon = (lat_max+lat_min)/2, (lon_max+lon_min)/2
# print(mid_lat)
# print(mid_lon)

# Create Folium Map
map = folium.Map(location=[mid_lat,mid_lon], zoom_start=11.5, control_scale=True,zoom_control=True,
               scrollWheelZoom=True,
               dragging=True)
# Add Tilelayer on map
folium.TileLayer('cartodbpositron').add_to(map)

# Add Markers
for i in range(len(lat)):
    folium.Circle([lat[i], lon[i]],
                  radius=60,
                  color=getColor((df.loc[i,'Sentiment Score']+1)*50),
                  fill_color=getColor((df.loc[i,'Sentiment Score']+1)*50),
                  fill=True,
                  fill_opacity=1).add_to(map)

# Show Map
map
```

```python
df = pd.read_csv('./final datasets/Silchar_Data.csv')

lat = df['Latitude']
lon = df['Longitude']

# Get nodes extent
margin = 0.0 # buffer to add to the range
lat_min = min(lat) - margin
lat_max = max(lat) + margin
lon_min = min(lon) - margin
lon_max = max(lon) + margin

# Get centre coordinates for map
mid_lat,mid_lon = (lat_max+lat_min)/2, (lon_max+lon_min)/2
# print(mid_lat)
# print(mid_lon)

# Create Folium Map
map = folium.Map(location=[mid_lat,mid_lon], zoom_start=12.5, control_scale=False,zoom_control=False,
               scrollWheelZoom=False,
               dragging=False)
# Add Tilelayer on map
folium.TileLayer('cartodbpositron').add_to(map)

# Add Markers
for i in range(len(lat)):
    folium.Circle([lat[i], lon[i]],
                  radius=20,
                  color="#000000").add_to(map)

# Show Map
map
```
