```python
import pandas as pd
import folium
import math
import numpy as np
import math
```

```python
def getColor(x):
    if x > 0:
        return '#009588'
    elif x == 0:
        return '#FFC73D'
    else:
        return '#AD1B42'
def getSize(x):
    if x==0:
        return 0
    else:
        return (x*3) + 40
def getDist(curr_coord,key,lat,lon):
    if math.dist(list(key),[lat,lon])<math.dist(list(curr_coord),[lat,lon]):
        return 1
    else:
        return -1
```

```python
df = pd.read_csv('./final datasets/Guwahati_Data.csv')

lat = df['Latitude']
lon = df['Longitude']

# print(lat)
# print(lon)

# Get nodes extent
margin = 0.01  # buffer to add to the range
lat_min = min(lat) - margin
lat_max = max(lat) + margin
lon_min = min(lon) - margin
lon_max = max(lon) + margin
# print(lat_min)
# print(lat_max)
# print(lon_min)
# print(lon_max)
# print("-------")
```

```python
matrix = {}
prev_lat,prev_lon=0,0
for i in np.arange(lat_min,lat_max,0.001):
    if i==lat_min:
        prev_lat=i
        continue
    for j in np.arange(lon_min,lon_max,0.001):
        if j==lon_min:
            prev_lon=j
            continue
        matrix[(((i-prev_lat)/2)+prev_lat,((j-prev_lon)/2)+prev_lon)] = (0,0.0)
        prev_lon=j
    prev_lat=i
    
print(len(matrix))

# Create dict for closest matrix point for each coordinate in csv
closest = {}

print("Closest Point calculation.....")

for i in range(len(lat)):
    closest[lat[i],lon[i]] = (0,0)
    curr_coord = closest[lat[i],lon[i]]
    for key in matrix.keys():
        if(getDist(curr_coord,key,lat[i],lon[i])>0):
            curr_coord = key
    closest[lat[i],lon[i]] = curr_coord
    tup = list(matrix[curr_coord])
    tup[0]+=1
    tup[1]+= df.loc[i, 'Sentiment Score']
    matrix[curr_coord]=tuple(tup)
    print(i,"done")
```

```python
np.save("one_closest.npy",closest)
np.save("one_matrix.npy",matrix)
```

```python
closest = np.load('one_closest.npy',allow_pickle='TRUE').item()
print(len(closest))
matrix = np.load('one_matrix.npy',allow_pickle='TRUE').item()
print(len(matrix))
```

```python
# Get centre coordinates for map
mid_lat, mid_lon = (lat_max+lat_min)/2, (lon_max+lon_min)/2
# print(mid_lat)
# print(mid_lon)

# Create Folium Map
map = folium.Map(location=[mid_lat, mid_lon], zoom_start=11.5, control_scale=False, zoom_control=False,
                 scrollWheelZoom=False,
                 dragging=False)
# Add Tilelayer on map
folium.TileLayer('cartodbpositron').add_to(map)

# Add Markers
for i in matrix.keys():
    if matrix[i][0]==0:
        continue
#     print(matrix[i])
    folium.Circle([i[0], i[1]],
                  radius=getSize(matrix[i][0]),
                  color=getColor(matrix[i][1]/matrix[i][0]),
                  fill_color=getColor(matrix[i][1]/matrix[i][0]),
                  fill=True,
                  fill_opacity=1).add_to(map)

# Show Map
map
```
