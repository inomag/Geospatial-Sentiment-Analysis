## WORD CLOUD BASED ON GUWAHATI DATASETS

```python
import pandas as pd
df = pd.read_csv(r'../final datasets/Guwahati_Data.csv', encoding ="latin-1")
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
full_text = ''
stopwords = set(STOPWORDS)

for val in df['Full Text']:
    val = str(val)
    val = val.replace("posted", "")
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    full_text += " ".join(tokens)+" "
```

```python
# full_text = ' '.join(dict.fromkeys(full_text.split()))
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(full_text)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
